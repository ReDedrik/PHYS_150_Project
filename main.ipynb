# Fractal of a Leaf Code
This fractal is called "Barnsley's Fern", named after Michael Barnsley, who founded the fractal.

```python
# Barnsley's Fern
x, y = 0, 0
x_list, y_list = [x], [y]
for i in range(100000):
    r = random.random()
    x_old, y_old = x, y
    if r < 0.01:
        x = 0
        y = 0.16 * y_old
    elif r < 0.86:
        x = 0.85 * x_old + 0.04 * y_old
        y = -0.04 * x_old + 0.85 * y_old + 1.6
    elif r < 0.93:
        x = 0.2 * x_old - 0.26 * y_old
        y = 0.23 * x_old + 0.22 * y_old + 1.6
    else:
        x = -0.15 * x_old + 0.28 * y_old
        y = 0.26 * x_old + 0.24 * y_old + 0.44
    x_list.append(x)
    y_list.append(y)
plt.scatter(x_list, y_list, s= 0.1, c=[0.2, 0.6, 0.2])
```

# Mandelbrot Set

```python
max_iterations = 1000
x_list, y_list = [], []
for i in range(600):
    for j in range(600):
        x = i / 200
        y = j / 200
        iterations, z = 0, 0
        
        while (abs(math.sqrt(z.real ** 2 + z.imag ** 2)) <= 2) and (iterations < max_iterations):
            z = z**2 + complex(x, y)
            iterations += 1
        if iterations >= 999:
            x_list.append(z.real)
            y_list.append(z.imag)
plt.scatter(x_list, y_list, s=0.1)

```
